```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python

df = pd.read_csv(r"C:\Users\HP\Desktop\gsfc\Sem6\Internship\K_ALGOS\Mall_Customers.csv")
print(df.head())
```

```python
# Drop 'CustomerID' and optionally 'Gender'
df_clean = df.drop(columns=['CustomerID', 'Gender'])
# Or encode Gender if you want:
# df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
# df_clean = df.drop(columns='CustomerID')
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clean)
```

```python
inertia = []

for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='Set2', s=100)
plt.title("Customer Segmentation using K-Means")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1–100)")
plt.legend()
plt.show()
```

```python
cluster_summary = df.groupby('Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
print(cluster_summary)
```
